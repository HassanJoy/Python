```python
def initial_course():
    rows, cols = int(input("enter initial course want to store: ")), 4
    course = []
    print(course)
    for i in range(rows):
        print("\nEnter cont %d details in the following order (ONLY):" % (i+1))
        temp = []
        for j in range(cols):
                     
            if j == 0:
                temp.append(str(input("Enter course code*: ")))
                
                if temp[j] == '' or temp[j] == ' ':
                                 print("mandatory")
                     
            if j == 1:
                temp.append(str(input("Enter course title: ")))
                if temp[j] == '' or temp[j] == ' ':
                
                    print("mandatory")
            if j == 2:
                temp.append(
                    str(input("Enter course credit: ")))
                
                if temp[j] == "" or temp[j] == ' ':
                   print("mandatory")
            if j == 3:
                temp.append(
                    str(input("Enter prerequisites: ")))
               
                if temp[j] == "" or temp[j] == ' ':
                   print("mandatory")          
        course.append(temp)
    print(course)
    return course
    
def menu():
    
    print("******************************************************************")
    print("\t\t\t options list", flush=False)
    print("******************************************************************")
    print("1. Add new course")
    print("2. Remove course")
    print("3. Delete all To list")
    print("4. Search course")
    print("5. Display courses")
    print("6. Store all course")
    print("7. Quit")
    choice = int(input("Please enter your option: "))
    return choice
 
def add_course(data):
    courses = []
    for i in range(len(data[0])):
        if i == 0:
            courses.append(str(input("Enter course code: ")))

        if i == 1:
            courses.append(str(input("Enter course title: ")))
        if i == 2:
            courses.append(
                str(input("Enter course credit: ")))
        if i == 3:
            courses.append(
                str(input("Enter prerequisites: ")))    
    data.append(courses)
    print("course added")
    return data
    
def remove_course(data):
    query = str( input("Please enter the name of the todo you wish to remove: "))
    temp = 0
    for i in range(len(data)):
        if query == data[i][0]:
            temp += 1
            print(data.pop(i))
            print("This course has been removed")
            return data
    if temp == 0:
        print("Sorry, this is not available in your list")
        return data
def delete_all(data):
    return data.clear()

def store(data):
  file = open("t.txt", "w")
 
#convert variable to string
  str = repr(data)
  file.write("input_dictionary = " + str + "\n")
 
#close file
  return data
  file.close()
   
 
def search_course(data):
    choice = int(input("Enter search criteria\n1. code\n2. tittle\n3.credit\n4.prerequisites\nPlease enter: "))
    temp = []
    check = -1
     
    if choice == 1:
    
        query = str( input("Please enter the name of the code you wish to search: "))
        for i in range(len(data)):
            if query == data[i][0]:
                check = i
                temp.append(data[i])
                 
    elif choice == 2:
   
        query = str(input("Please enter the tittle\ of the contact you wish to search: "))
        for i in range(len(data)):
            if query == data[i][1]:
                check = i
                temp.append(data[i])
                 
    elif choice == 3:
   
        query = str( input("Please enter the credit of the to do you wish to search: "))
        for i in range(len(data)):
            if query == data[i][2]:
                check = i
                temp.append(data[i])
         
    else:
        print("Invalid search criteria")
        return -1
    if check == -1:
        return -1
    else:
        display_course(temp)
        return check

def display_courses(data):
    if not data:
        print("List is empty: []")
    else:
        for i in range(len(data)):
            print(data[i])
def thanks():

    print("******************************************************************")
    print("Thank you for using our TO DO system.")
    print("Please visit again!")
    print("******************************************************************")
    print("Goodbye, have a nice day ahead!")
    
print("....................................................................")
print("Hello dear user, welcome to our TO DO system")
print("You may now proceed to explore this directory")
print("....................................................................")
ch = 1
data= initial_course()
while ch in (1, 2, 3, 4, 5):
    ch = menu()
    if ch == 1:
        data = add_course(data)
    elif ch == 2:
        data = remove_course(data)
    elif ch == 3:
        data = delete_all(data)
    elif ch == 4:
        data = search_course(data)
        if data == -1:
            print("The contact does not exist. Please try again")
    elif ch == 5:
        display_courses(data)
    elif ch == 6 :
        store(data)    
    else:
        thanks()
```
