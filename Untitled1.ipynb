```python
def menu():
    
    print("******************************************************************")
    print("\t\t\tTODO LIST", flush=False)
    print("******************************************************************")
    #print("\tYou can now perform the following operations on this phonebook\n")
    print("1. Add new course")
    print("2. Remove course")
    print("3. Delete all TO DO")
    print("4. Search course")
    print("5. Display all course")
    print("6. Store all course")
    print("7. Exit todo")
    choice = int(input("Please enter your choice: "))
    return choice
```

```python
def add_contact(pb):
    dip = []
    for i in range(len(pb[0])):
        if i == 0:
            dip.append(str(input("Enter course code: ")))
        
        if i == 1:
            dip.append(str(input("Enter course title: ")))
        if i == 2:
            dip.append(
                str(input("Enter course credit: ")))
        if i == 3:
            dip.append(
                str(input("Enter prerequisites: ")))    
    pb.append(dip)
    # And once you modify the list, you return it to the calling function wiz main, here.
    return pb
```

```python
def remove_existing(pb):
    query = str(
        input("Please enter the name of the todo you wish to remove: "))
    temp = 0
    for i in range(len(pb)):
        if query == pb[i][0]:
            temp += 1
            print(pb.pop(i))
            print("This query has now been removed")
            return pb
    if temp == 0:
        print("Sorry, you have entered an invalid query.\
    Please recheck and try again later.")
         
        return pb
```

```python
def store(pb):
  
  file = open("t.txt", "w")
 
#convert variable to string
  str = repr(pb)
  file.write("input_dictionary = " + str + "\n")
 
#close file
  return pb
  file.close()
   
 
```

```python

def search_existing(pb):
    choice = int(input("Enter search criteria\n1. code\n2. tittle\n3.credit\n4.prerequisites\nPlease enter: "))
    temp = []
    check = -1
     
    if choice == 1:
    # This will execute for searches based on contact name
        query = str(
            input("Please enter the name of the code you wish to search: "))
        for i in range(len(pb)):
            if query == pb[i][0]:
                check = i
                temp.append(pb[i])
                 
    elif choice == 2:
    # This will execute for searches based on todo''s date of birth
        query = str(input("Please enter the tittle\
            of the contact you wish to search: "))
        for i in range(len(pb)):
            if query == pb[i][1]:
                check = i
                temp.append(pb[i])
                 
    elif choice == 3:
    # This will execute for searches based on contact category
        query = str(
            input("Please enter the credit of the to do you wish to search: "))
        for i in range(len(pb)):
            if query == pb[i][2]:
                check = i
                temp.append(pb[i])
        # All contacts under query category will be shown using this feature
         
    else:
    # If the user enters any other choice then the search will be unsuccessful
        print("Invalid search criteria")
        return -1
    # returning -1 indicates that the search was unsuccessful
     
    # all the searches are stored in temp and all the results will be displayed with
    # the help of display function
     
    if check == -1:
        return -1
        # returning -1 indicates that the query did not exist in the directory
    else:
        display_all(temp)
        return check
        # we're just returning a index value wiz not -1 to calling function just to notify
        # that the search worked successfully
 
# this function displays all content of phonebook pb
def display_all(pb):
    if not pb:
    # if display function is called after deleting all contacts then the len will be 0
    # And then without this condition it will throw an error
        print("List is empty: []")
    else:
        for i in range(len(pb)):
            print(pb[i])
 
```
