```python
|#import numpy library
import numpy as np
plt.style.use('classic')
#check avilable style in matplotlib
print(plt.style.available)
```

```python
#create an numpy array 0 to 10 with 100 items divided 
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x),'-')
plt.plot(x,np.cos(x),'--')
plt.show()
```

```python
fig=plt.figure()
plt.plot(x,np.sin(x),'-')
plt.plot(x,np.cos(x),'--');
```
